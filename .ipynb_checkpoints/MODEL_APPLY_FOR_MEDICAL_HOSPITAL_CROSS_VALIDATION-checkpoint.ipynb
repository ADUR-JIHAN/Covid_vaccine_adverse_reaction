# SPLIT X AND Y VALUES FOR MEDICAL HISTORIES (HOSPITAL)

```python
clm = new_df_with_medical_history.columns
feature_column_names = clm[:-3]

predicted_class_name = ['HOSPITAL']

X = new_df_with_medical_history[feature_column_names].values
y = new_df_with_medical_history[predicted_class_name].values

```

# SMOTING 

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE()
X, y = sm.fit_resample(X, y) 

```

# SPLIT THE TRAIN AND TEST DATA

```python
split_test_size = 0.20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_test_size, random_state = 42)

print("{0:0.2f}% in training set".format((len(X_train)/len(new_df_with_medical_history.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(new_df_with_medical_history.index)) * 100))

```

# STANDARD SCALLING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# APPLY MODEL

```python
# evaluate bagging algorithm for classification
from numpy import mean
from numpy import std
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import sklearn.metrics as metrics
import sklearn.metrics as metrics
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn import model_selection
def autoML(algorithm): 
    skfold = StratifiedKFold(n_splits=3, random_state=100,shuffle=True)
    model_skfold = algorithm()
    results_skfold = model_selection.cross_val_score(model_skfold, X, y, cv=skfold)
    print(f"----------------------------------------------------------------------------\n")
    print(f"Model Name: {algorithm.__name__}\nTrain Accuracy: {results_skfold.mean()*100.0}\n\n")
    model = algorithm().fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print ("Confusion Matrix ")
    # labels for set 1=True to upper left and 0 = False to lower right
    print ("{0}".format(metrics.confusion_matrix(y_test, y_pred, labels=[1, 0])))
    print ("Classification Report\n")
    # labels for set 1=True to upper left and 0 = False to lower right
    print ("{0}".format(metrics.classification_report(y_test, y_pred, labels=[1, 0])))
    y_pred = np.where(y_pred > 0.5, 1, 0)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    test_acc = metrics.accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)
    logloss = log_loss(y_test, y_pred)
    print('Test Accuracies')
    print ("{0:.2f}".format(test_acc))
    print(f"precision: {precision}\n")
    print(f"recall: {recall}\n")    
    print(f"f1: {f1}\n")
    print(f"auc: {auc}\n")
    print(f"logloss: {logloss}\n")
    print(f"----------------------------------------------------------------------------\n") 
    print ("precision:{0:.2f}".format(precision))
    print ("recall:{0:.2f}".format(recall))
    print ("f1:{0:.2f}".format(f1))
    print ("auc:{0:.2f}".format(auc))
    print ("logloss:{0:.2f}".format(logloss))
    
    
```

```python
from sklearn.model_selection import KFold, cross_val_score
import optuna
```

```python
RANDOM_SEED = 42

# 10-fold CV
kfolds = KFold(n_splits=10, shuffle=True, random_state=RANDOM_SEED)
```

```python
def tune(objective):
    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=100)

    params = study.best_params
    best_score = study.best_value
    print(f"Best score: {best_score} \nOptimized parameters: {params}")
    return params
```

```python
import optuna
import sklearn
from sklearn import datasets
def objective(trial):
    
    # parameters to optimize
    
    n_estimators = trial.suggest_int('n_estimators', 20, 40, 1)
    random_state = trial.suggest_int('random_state', 40, 45, 1)
    max_depth = trial.suggest_int('max_depth', 1, 32)
    max_features = trial.suggest_int('max_features', 2, 13)
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])
    criterion = trial.suggest_categorical('criterion', ['entropy','gini'])
   # min_samples_split = trial.suggest_discrete_uniform('min_samples_split', 0.1, 1.0, 0.1)
   # min_samples_leaf = trial.suggest_discrete_uniform('min_samples_leaf', 0.1, 0.5, 0.1)
    
    # our ML pipeline
    clf = RandomForestClassifier(n_estimators=n_estimators,
                                                                max_depth=max_depth,
                                                                max_features=max_features,
                                                                bootstrap=bootstrap,
                                                                random_state= random_state
                                                                )
    
    # return average 5-fold CV accuracy
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    return cross_val_score(clf, X_test, y_test, scoring='accuracy', n_jobs=-1, cv=cv,error_score='raise').mean()
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=3000)
```

```python
trial = study.best_trial
print('Accuracy: {}'.format(trial.value))

print('best parameter: {}'.format(trial.params))
```
